## ¬øQu√© es una gu√≠a de estilo en programaci√≥n?

Una gu√≠a de estilo es un conjunto de reglas de escritura de c√≥digo que un programador, o un equipo de ellos, sigue para escribir el c√≥digo de un programa. Esto asegura que todo archivo escrito siga el mismo formato y parezca provenir de un mismo autor. Estas reglas puede ser variadas: desde **reglas de formato** que deciden si llamar a una variable `mivariable`, `mi_variable` o `miVariable`; hasta **reglas sem√°nticas** que deciden si llamar a una funci√≥n que suma elementos de una lista `funcion_que_suma` o simplemente `suma`. Qu√© y cu√°ntas reglas se usan en un proyecto queda completamente definido a gusto de los colaboradores. En general, al trabajar en un equipo grande, se definen en conjunto estas reglas antes de comenzar el proyecto o se usan reglas previamente creadas.

## ¬øPor qu√© usar una gu√≠a de estilo?

A primera vista, el seguir este tipo de reglas puede verse como una restricci√≥n o m√°s trabajo a la hora de programar. Sin embargo, el escribir c√≥digo conscientemente siguiendo una gu√≠a de estilo trae muchas ventajas. Espec√≠ficamente, trae ahorros de tiempo considerables tanto a la hora de escribir c√≥digo como posterior a su escritura. Es por esto que su uso es una metodolog√≠a de desarrollo popular en la industria. Distintos equipos de desarrollo, en lugares como Facebook, Google, Platanus, Fintual, Cornershop y m√°s, utilizan diversas gu√≠as de estilo, dependiendo de los proyectos y lenguajes utilizados. 

A continuaci√≥n se listan algunas ventajas concretas de usar gu√≠as de estilo:

### 1. El c√≥digo es m√°s le√≠ble

El tener reglas de escritura de variables, funciones y clases permite que el c√≥digo completo sea m√°s f√°cil de leer. Al programar, uno pasa m√°s tiempo leyendo c√≥digo que escribi√©ndolo, por lo que es fundamental contar con c√≥digo legible que nos permita entender bien o que alguien m√°s entienda lo que ocurre en el programa. Python espec√≠ficamente, tiene la gracia de una sintaxis muy legible, pero esta propiedad puede perderse r√°pido con malos h√°bitos de escritura.

### 2. Hace el *debuggeo* m√°s r√°pido

Siguiendo la idea anterior: como programador, uno tiende a quedarse mirando el c√≥digo por largos periodos de tiempo, sobre todo cuando queremos encontrar un error. A esta actividad de encontrar un problema o un *bug* (üêõ) en un programa se le conoce como *debugging*. Si nuestro c√≥digo es ordenado y le√≠ble, entonces nos puede ahorrar tiempo en encontrar *bugs*.

### 3. Se escribe c√≥digo m√°s r√°pido

A veces, puede tomar tiempo decidir cu√°l es el nombre m√°s apropiado para cada elemento (variables, funciones y clases). El tener reglas definidas para nombrar objetos nos ayuda a tomar menos decisiones, y por lo tanto, nombrar y escribir m√°s r√°pido.

### 4. Nombres de archivos y variables predecibles

En proyectos grandes, lo m√°s seguro es que se usar√°n m√∫ltiples archivos, clases y funciones escritas por ti o por otra persona. Al tener que reutilizar dicho c√≥digo, es dif√≠cil recordar el nombre de cada funci√≥n, m√©todo y clase escrito, lo que provoca perder tiempo buscando dichos nombres manualmente. El tener reglas para escribir y nombrar archivos y variables nos puede ayudar en ahorrar tiempo en esa b√∫squeda.

## Gu√≠a de estilo en Python: PEP8

PEP8 es la gu√≠a de estilo propuesta por la documentaci√≥n oficial de Python. Esta gu√≠a define muchas reglas que siguen los pilares de la filosof√≠a de los creadores de Python. Puedes encontrar la especificaci√≥n completa [aqu√≠](https://www.python.org/dev/peps/pep-0008/). 

En IIC2233 utilizaremos solo un subconjunto de reglas de PEP8 para escribir nuestro c√≥digo (y esperamos que los estudiantes tambi√©n lo usen en sus tareas). Estas reglas son las siguientes:

### - Aplicar `snake_case` para nombres de variables, atributos, funciones y m√©todos.

`snake_case` es la pr√°ctica de escribir palabras o frases compuestas usando `_` para separar los distintos elementos del nombre y sin espacios. Existen variantes que utilizan may√∫sculas para la primera letra, pero en IIC2233 solo usamos min√∫sculas para este tipo de nombres, y lo usaremos para nombrar **pr√°cticamente todo**, menos los nombres de clases. Esto es: **nombres de variables**, **atributos**, **funciones** y **m√©todos**.

### - Aplicar `CamelCase` para nombres de clases

Por el otro lado, `CamelCase` es la pr√°ctica de escribir palabras o frases compuestas sin espacios ni separador entre elementos, y por cada elemento se comienza con una may√∫scula. Existen variantes donde la primera letra utilizan min√∫scula (*lower cammel case*), pero en IIC2233 utilizaremos la opuesta: *upper cammel case*. Este tipo de escritura lo reservamos para el **nombre de clases**.

A continuaci√≥n, un ejemplo de escritura siguiendo `snake_case` y `CamelCase`:

```python
class PerroAmigable:

    def __init__(self, nombre, edad):
        self.nombre = nombre
        self.edad = edad

    def saludar_amigo(self, amigo):
        print("¬°Woof!")
        print("¬°Hola " + amigo.nombre + "!")


def saludo_mutuo(perro_1, perro_2):
    perro_1.saludar_amigo(perro_2)
    perro_2.saludar_amigo(perro_1)
    print()


perro_juan = PerroAmigable("Juan", 21)
perro_pedro = PerroAmigable("Pedro", 20)
perro_jose = PerroAmigable("Jose", 22)

saludo_mutuo(perro_juan, perro_pedro)
saludo_mutuo(perro_jose, perro_pedro)
saludo_mutuo(perro_juan, perro_jose)
```

### - Nombres de variables descriptivos.

Es usual que al programar se nos ocurra nombres de variables simples como `a` o `variable`. Si bien tal vez esto ahorra en tiempo de escritura, esto no genera c√≥digo le√≠ble o f√°cil de entender. Al volver a leer c√≥digo como el siguiente, nos produce dudas de que realmente hace lo que estamos leyendo:

```python
def funcion(l):
    i = 0
    a = l[0]
    for j in range(1, len(l)):
        var = l[j]
        if var > a:
            j = i
            a = var
    return i
```

En cambio, al utilizar nombres m√°s descriptivos, queda m√°s claro lo que hace la funci√≥n:

```python
def indice_de_maximo(lista):
    indice_actual = 0
    maximo_actual = lista[0]
    for indice in range(1, len(lista)):
        elemento = lista[indice]
        if elemento > maximo_actual:
            indice_actual = indice
            maximo_actual = elemento
    return indice_actual
```

### - Un espacio despu√©s de ‚Äú,‚Äù.

Para mantener orden, formato y legibilidad, esta regla se aplica en todos los casos que se usan comas al enumerar elementos:

```python
from random import randint, choice  # Espacios al importar 


def operaciones_con_potencia(lista, potencia):  # Espacios en definici√≥n de funciones
    suma = 0
    multiplicacion = 1
    for elemento in lista:
        suma += elemento ** potencia
        multiplicacion *= elemento ** potencia
    return suma, multiplicacion  # Espacios en retornos


numeros = [3, 6, 2, 4]  # Espacios en definici√≥n de listas
operaciones_con_potencia(numeros, 2)  # Espacios en llamadas de funciones
```

### - L√≠neas de largo m√°ximo de 100 caracteres (incluyendo espacios).

Es recurrente el problema de l√≠neas muy largas, sobre todo en l√≠neas ya indentadas. Generalmente, se establece el l√≠mite de 80 caracteres como m√°ximo para seguir una convenci√≥n de compatibilidad con herramientas de edici√≥n que establecen este l√≠mite y permiten ver el c√≥digo solo de ese ancho de espacio. Pero, en el curso relajaremos esta regla y trabajaremos con l√≠neas de **m√°ximo 100 caracteres**. Hay varias formas de evitar este problema, pero depende del contexto en el que se encuentre el problema. A veces es conveniente aprovechar la creaci√≥n de variables extras para expresiones largas y acortar l√≠neas, o en casos extremos, el uso de `\` que permite separar una l√≠nea larga en la siguiente.

Por ejemplo, el siguiente c√≥digo tiene l√≠neas que se pueden acortar:

```python
print("Debe ingresar dos n√∫meros")
valor_1 = int(input("Ingrese primer n√∫mero"))
valor_2 = int(input("Ingrese segundo n√∫mero"))
# La siguiente l√≠nea es muy larga
if valor_1 + valor_2 >= 1000 and (valor_1 - valor_2 >= 1000 or valor_2 - valor_1 >= 1000):
    print("Los n√∫meros suman mucho y est√°n muy separados")
elif valor_1 + valor_2 >= 1000:
    print("Los n√∫meros suman mucho")
elif (valor_1 - valor_2 >= 1000 or valor_2 - valor_1 >= 1000):
    print("Los n√∫meros est√°n muy separados")
else:
    print("Los n√∫meros son apropiados")
```

Mediante el uso de `\`:

```python
print("Debe ingresar dos n√∫meros")
valor_1 = int(input("Ingrese primer n√∫mero"))
valor_2 = int(input("Ingrese segundo n√∫mero"))
# Se separ√≥ usando \
if valor_1 + valor_2 >= 1000 and \
            (valor_1 - valor_2 >= 1000 or valor_2 - valor_1 >= 1000):
    print("Los n√∫meros suman mucho y est√°n muy separados")
elif valor_1 + valor_2 >= 1000:
    print("Los n√∫meros suman mucho")
elif (valor_1 - valor_2 >= 1000 or valor_2 - valor_1 >= 1000):
    print("Los n√∫meros est√°n muy separados")
else:
    print("Los n√∫meros son apropiados")
```

O el uso de variables auxiliares:

```python
print("Debe ingresar dos n√∫meros")
valor_1 = int(input("Ingrese primer n√∫mero"))
valor_2 = int(input("Ingrese segundo n√∫mero"))
# Variables auxiliares para mejorar legibilidad
suma = valor_1 + valor_2
diferencia = abs(valor_1 - valor_2)
if suma >= 1000 and diferencia >= 1000:
    print("Los n√∫meros suman mucho y est√°n muy separados")
elif suma >= 1000:
    print("Los n√∫meros suman mucho")
elif diferencia >= 1000:
    print("Los n√∫meros est√°n muy separados")
else:
    print("Los n√∫meros son apropiados")
```
