### Ejercicio 1: Ordenar

Confeccione un servidor que permita ordenar una lista de números de mayor a menor. No obstante, este servidor tiene un comportamiento peculiar. Si le llega el mensaje `"reverse"`, desde ahora ordenará todas las listas en la dirección opuesta, es decir, de menor a mayor. Y solo volverá a la normalidad si vuelve a recibir el mensaje `"reverse"`.

Para lograr esto, confeccione un servidor que tenga 2 comportamientos posibles. 

1. Si llega el mensaje `"reverse"`, entonces el modo del servidor cambia y ahora, cada vez que ordene una lista, lo hará de forma inversa a lo que estaba antes. En este sentido, si el servidor ordena de menor a mayor, ahora ordenará de mayor a menor. Pero si ya ordena de mayor a menor y llega este mensaje, entonces ahora ordenará de menor a mayor.

2. Si le llega un mensaje con el siguiente formato :: En caso que no llegue el mensaje "reverse", entonces está llegando un _string_ que consiste en la concatenación, mediante comas, de varios números. Por ejemplo: `"1,32,65,22,4444,11"`. En este caso, el servidor retornará la secuencia ordenada según el modo en el que está (mayor a menor o menor a mayor). 


Finalmente, confecciona una interfaz que contenga 4 elementos:
- Un _input_ de texto (`QLineEdit`) para poner la lista de números del siguiete modo: A,B,C,D,E
- Un _output_ de texto (`QLabel`) para mostrar la respuesta del servidor.
- Un botón que diga `Ordenar` que cuando es oprimido, se manda al servidor la lista contenida en el _input_ de texto, y luego se actualiza el _output_ con la lista ordenada.
- Un botón que diga `reverse` que cuando es oprimido, manda el mensaje "reverse" al servidor.

### Ejercicio 2: Adivina el número

Te pidieron confeccionar un pequeño juego de adivinar el número que tenga el servidor. Este juego se puede jugar entre múltiples jugadores y las reglas son estas:

1. El servidor tiene definido, en una constante de su código, el número a adivinar.
2. Si un jugador adivina el número, le responderá con un mensaje de felicitaciones. Luego, a los demás jugadores les mandará un mensaje que obligará, al cliente, a cerrar el programa.
3. Si un jugador no adivina el número o manda algo que no es un número, se le dirá que está equivocado y el servidor no le permitirá adivinar nuevamente hasta pasado 3 segundos. En este sentido, el cliente puede mandar mensajes, pero el servidor los va a ignorar hasta que pasen 3 segundos. Luego de ese tiempo, va a procesar los nuevos mensajes que lleguen.
4. Si un jugador falla 5 veces, el servidor le mandará un mensaje que obligará, al cliente, a cerrar el programa.

Adicionalmente, deberás hacer una pequeña interfaz gráfica que contenga 3 elementos:
- Un _input_ de texto (`QLineEdit`) para poner el número a adivinar.
- Un _output_ de texto (`QLabel`) para mostrar la respuesta del servidor.
- Un botón que diga `Enviar número` que cuando es oprimido, se manda al servidor el número contenido en el _input_ de texto.
